```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("results.xlsx")
df.head(10)
```

```python
def makePlot(matrixSize, df):
    df_temp = df[['Blocks', matrixSize]]
    return df_temp.set_index('Blocks').plot(kind='bar', figsize=(10,6), colormap='Accent')
```

```python
sizes = [512, 1024, 1536, 2048]
for size in sizes:
    makePlot(size, df)
    plt.savefig(f'plot{size}.png', dpi=300)
```
